# numpy类似于python中的列表，他们都可以作为容器有增删改查的功能
# 在一维以上维度是向量化操作比列表快，打在末尾添加元素比列表慢



```python
import numpy as np
# 向量初始化
a=np.array([1.,2.,3.])
a.dtype
a.shape
```

```python
# 用np.zeros和np.empty初始化数组，预分配必要的空间
b=np.zeros(3,int)
b
```

```python
# 创建一个空数组，大小和元素类型与现有数组相同
c=np.zeros_like(a)
c
```

```python
# 用常量的函数填充基本都有_like
d=np.ones_like(a)
d
```

```python
d=np.full_like(a,8)
d
```

```python
# 在npmpy中，可以用arrange或者linspace来初始化单调序列数组
np.arange(6)
np.arange(2,6)
np.arange(1,6,2)
```

```python
np.linspace(0,0.5,7)
```

```python
# 向量索引
# 基础的向量索引操作，只是展示部分数据，而不改变数据本身
import numpy as np
a=np.arange(1,6)
a[0:2]
```

```python
# 布尔操作
a=np.arange(10)
print(np.any(a>5))
print(a[a>5])
print(np.all(a>5))
# 也可用where和clip代替上面的方法
print(np.where(a>5))
print(np.nonzero(a>5))
print(np.where(a>=5,1,0))
print(np.clip(a,3,5))
```
